```python
import networkx as nx
import matplotlib.pyplot as plt

graph = {
    'A': [('B', 4), ('C', 2)],
    'B': [('A', 4), ('D', 5)],
    'C': [('A', 2), ('D', 1)],
    'D': [('B', 5), ('C', 1)]
}

coordinates = {
    'A': (0, 0),
    'B': (1, 2),
    'C': (2, 0),
    'D': (3, 2)
}

def draw_simple_graph(graph, coords, path=None):
    G = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor, weight in neighbors:
            G.add_edge(node, neighbor, weight=weight)

    edge_labels = nx.get_edge_attributes(G, 'weight')

    nx.draw(G, coords, with_labels=True, node_color='lightblue', node_size=700, edgecolors='black')
    nx.draw_networkx_edge_labels(G, coords, edge_labels=edge_labels, font_color='red')

    if path:
        nx.draw_networkx_nodes(G, coords, nodelist=path, node_color='orange')
        nx.draw_networkx_edges(G, coords, edgelist=list(zip(path, path[1:])), edge_color='orange', width=3)

    plt.title("Test Graph")
    plt.show()

draw_simple_graph(graph, coordinates, path=['A', 'C', 'D'])
```
